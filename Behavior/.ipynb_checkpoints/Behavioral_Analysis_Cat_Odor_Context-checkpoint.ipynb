```python
import numpy as np
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import ntpath as nt
import re

#Put file directory here
base_dir = os.path.abspath('.')


```

```python
safecues = np.array(range(0, 12))
threatcues = np.array(range(12, 24))


safe_foodarea= pd.DataFrame()
threat_foodarea = pd.DataFrame()

safe_hiddenarea= pd.DataFrame()
threat_hiddenarea = pd.DataFrame()

safe_latency= pd.DataFrame()
threat_latency= pd.DataFrame()

safe_rewarded= pd.DataFrame()
threat_rewarded= pd.DataFrame()

safe_freezing= pd.DataFrame()
threat_freezing= pd.DataFrame()

firstbarcue = pd.DataFrame()



files = glob.glob(base_dir + '*/*.csv')
files = files


for f in range(len(files)):
    
    a = pd.read_csv(files[f])
    
    
    cue = np.diff(a['Tone active'].values) #cue onsite and end
    cueon = a[a['Tone active'].values > 0] #cue on subset
    time = a['Time'].values
    cuestart = np.where(cue>0)[0] + 1  #Cue onset index
    cueend = np.where(cue<0)[0] + 1   #Cue end index

    barincue = cueon[cueon['Bar Press active'].values > 0]
    bar = np.where(a['Bar Press active'].values > 0)[0]
    firstbar = np.array([], dtype = 'int64')
    todelete = np.array([], dtype = 'int64')
    firstbarindex = np.array([], dtype = 'int64')
    rewarded = np.array([])
    
    latency = time[cueend] - time[cuestart]
    
    safe_latency = safe_latency.append(pd.DataFrame(latency[:12]).T)
    threat_latency = threat_latency.append(pd.DataFrame(latency[12:]).T)

    #first bar press during each cue
    for n in range(len(cuestart)):
        diff = bar - cuestart[n]
        if len(diff[diff>0]) == 0:
            firstbarindex = len(a)
        else:
            firstbarindex = bar[list(diff).index(min(diff[diff>0]))]
        if firstbarindex > cueend[n]:
            firstbar = np.append(firstbar, 0)
        else:
            firstbar = np.append(firstbar, firstbarindex)
    
    firstbarcue = firstbarcue.append(pd.DataFrame(firstbar).T)
    safe_rewarded = firstbarcue.iloc[:,:12]>0
    threat_rewarded = firstbarcue.iloc[:,12:]>0
    
    #15s after cue
    fifteenaftercue = np.array([], dtype = 'int64')
    cueperiod = np.array([])
    for i in range(0,len(cuestart)):
        diff = a['Time']-a.loc[cuestart[i],'Time']-15
        fifteenaftercue_index = np.abs(diff).argmin()
        fifteenaftercue = np.append(fifteenaftercue, fifteenaftercue_index)
        duration = time[fifteenaftercue[i]] - time[cuestart[i]]
        cueperiod = np.append(cueperiod, duration)
        
    fifteen_index_diff =  fifteenaftercue- cuestart
    
    #food area within 15s after cuestart
    foodarea = np.where(a['In Food Area'].values>0)[0]
    foodareastart = np.where(np.diff(a['In Food Area'].values,prepend = 0, append =0)>0)[0]
    foodareaend = np.where(np.diff(a['In Food Area'].values,prepend = 0, append =0)<0)[0]
    foodareaduration = np.array([])
    
    for n in range(len(cuestart)):
        diff = foodarea - cuestart[n]
        diff2 = foodareastart - cuestart[n]
        duration = np.array([0])
        if len(diff[diff>=0]) == 0:
            duration = 0
        elif foodarea[list(diff).index(min(diff[diff>=0]))]>fifteenaftercue[n]:
            duration = 0
        else:
            foodareastart_index = foodarea[list(diff).index(min(diff[diff>=0]))]
            foodareaindex_within = foodareastart[(foodareastart>foodareastart_index)&(foodareastart<=fifteenaftercue[n])]
            foodareastart_index = np.append(foodareastart_index, foodareaindex_within)
            
            for i in range(len(foodareastart_index)):
                foodareaend_index = foodareaend[foodareaend>foodareastart_index[i]].min()
                if foodareaend_index > fifteenaftercue[n]:
                    foodareaend_index = fifteenaftercue[n]
                period = time[foodareaend_index]-time[foodareastart_index[i]]
                duration = duration + period
        foodareaduration = np.append(foodareaduration, duration)
    
    
    safe_foodarea = safe_foodarea.append(pd.DataFrame(foodareaduration[:12]).T)
    threat_foodarea = threat_foodarea.append(pd.DataFrame(foodareaduration[12:]).T)

    #hidden area within 15s after cuestart
    hiddenarea = np.where(a['In Hidden Area'].values>0)[0]
    hiddenareastart = np.where(np.diff(a['In Hidden Area'].values,prepend = 0, append =0)>0)[0]
    hiddenareaend = np.where(np.diff(a['In Hidden Area'].values,prepend = 0, append =0)<0)[0]
    hiddenareaduration = np.array([])
    
    for n in range(len(cuestart)):
        diff = hiddenarea - cuestart[n]
        diff2 = foodareastart - cuestart[n]
        duration = np.array([0])
        if len(diff[diff>=0]) == 0:
            duration = 0
        elif hiddenarea[list(diff).index(min(diff[diff>=0]))]>fifteenaftercue[n]:
            duration = 0
        else:
            hiddenareastart_index = hiddenarea[list(diff).index(min(diff[diff>=0]))]
            hiddenareaindex_within = hiddenareastart[(hiddenareastart>hiddenareastart_index)&(hiddenareastart<=fifteenaftercue[n])]
            hiddenareastart_index = np.append(hiddenareastart_index, hiddenareaindex_within)
            
            for i in range(len(hiddenareastart_index)):
                hiddenareaend_index = hiddenareaend[hiddenareaend>hiddenareastart_index[i]].min()
                if hiddenareaend_index > fifteenaftercue[n]:
                    hiddenareaend_index = fifteenaftercue[n]
                period = time[hiddenareaend_index]-time[hiddenareastart_index[i]]
                duration = duration + period
        hiddenareaduration = np.append(hiddenareaduration, duration)
    
    
    safe_hiddenarea = safe_hiddenarea.append(pd.DataFrame(hiddenareaduration[:12]).T)
    threat_hiddenarea = threat_hiddenarea.append(pd.DataFrame(hiddenareaduration[12:]).T)

    #freezing within 15s after cuestart
    freezing = np.where(a['Freezing'].values>0)[0]
    freezingstart = np.where(np.diff(a['Freezing'].values,prepend = 0, append =0)>0)[0]
    freezingend = np.where(np.diff(a['Freezing'].values,prepend = 0, append =0)<0)[0]
    freezingduration = np.array([])
    
    for n in range(len(cuestart)):
        diff = freezing - cuestart[n]
        diff2 = freezingstart - cuestart[n]
        duration = np.array([0])
        if len(diff[diff>=0]) == 0:
            duration = 0
        elif freezing[list(diff).index(min(diff[diff>=0]))]>fifteenaftercue[n]:
            duration = 0
        else:
            freezingstart_index = freezing[list(diff).index(min(diff[diff>=0]))]
            freezingindex_within = freezingstart[(freezingstart>freezingstart_index)&(freezingstart<=fifteenaftercue[n])]
            freezingstart_index = np.append(freezingstart_index, freezingindex_within)
            
            for i in range(len(freezingstart_index)):
                freezingend_index = freezingend[freezingend>freezingstart_index[i]].min()
                if freezingend_index > fifteenaftercue[n]:
                    freezingend_index = fifteenaftercue[n]
                period = time[freezingend_index]-time[freezingstart_index[i]]
                duration = duration + period
        freezingduration = np.append(freezingduration, duration)
    
    
    safe_freezing = safe_freezing.append(pd.DataFrame(freezingduration[:12]).T)
    threat_freezing = threat_freezing.append(pd.DataFrame(freezingduration[12:]).T)

    
    
```

```python
rewardedcues = np.hstack([np.arange(11), 14])
omittedcues = np.hstack([np.arange(11,14), np.arange(15,23)])


omittedbar_exp = pd.DataFrame()
rewardedbar_exp = pd.DataFrame()
omittedburstpress_exp = pd.DataFrame()
rewardedburstpress_exp = pd.DataFrame()
omittedmaxspeed_exp = pd.DataFrame()
rewardedmaxspeed_exp = pd.DataFrame()
animal_id = np.array([])

files = glob.glob(expected_dir + '/*[Test]*.csv')

animal = pd.read_excel(animal_dir)

for f in range(len(files)):
    
    a = pd.read_csv(files[f])
    file = nt.basename(files[f])
    animal['File Name'].str.match(pat = file.replace('.csv', ''))
    animal_id = np.append(animal_id, animal.loc[animal['File Name'].str.match(pat = file.replace('.csv', '')), 'Animal ID'].values)
    
    cue = np.diff(a['Cue Light active'].values) #cue onsite and end
    cueon = a[a['Cue Light active'].values > 0] #cue on subset
    time = a['Time'].values
    cuestart = np.where(cue>0)[0] + 1  #Cue onset index
    cueend = np.where(cue<0)[0] + 1   #Cue end index

    barincue = cueon[cueon['Bar Press active'].values > 0]
    bar = np.where(a['Bar Press active'].values > 0)[0]
    firstbar = np.array([], dtype = 'int64')
    todelete = np.array([], dtype = 'int64')

    #first bar press during each cue
    for n in range(len(cuestart)):
        diff = bar - cuestart[n]
        if len(diff[diff>0]) == 0:
            firstbarindex = len(a)
        else:
            firstbarindex = bar[list(diff).index(min(diff[diff>0]))]
        if firstbarindex > cueend[n]:
            firstbar = np.append(firstbar, 0)
        else:
            firstbar = np.append(firstbar, firstbarindex)


    #15s after bar press
    barend = np.array([], dtype = 'int64')
    barperiod = np.array([])
    for i in range(0,len(firstbar)):
        if firstbar[i] == 0:
            barend = np.append(barend, 0)
            barperiod = np.append(barperiod, 0)
        else:
            diff = a['Time']-a.loc[firstbar[i],'Time']-15
            barendindex = np.abs(diff).argmin()
            barend = np.append(barend, barendindex)
            duration = time[barend[i]] - time[firstbar[i]]
            barperiod = np.append(barperiod, duration)


    #press frequency

    barpress = np.array([])
                       
    for i in range(len(firstbar)):
        if barend[i] == 0:
            barpress = np.append(barpress, 0)
        else:
            barpress = np.append(barpress, sum(np.diff(a.loc[firstbar[i]:barend[i], 'Bar Press active'].values)<0)/(barperiod[i]/60))

    omittedbar = barpress[omittedcues]
    rewardedbar = barpress[rewardedcues]

    #burst pressing
    burstpress = np.array([])
    for i in range(len(firstbar)):
        if barend[i] == 0:
            burstpress = np.append(burstpress, 0)
        else:
            b = a.loc[firstbar[i]:barend[i],:].copy()
            b.loc[:,'dB'] = b.loc[:,'Bar Press active'].diff(-1).copy()
            b.loc[:,'dB'] = b.loc[:,'dB'].mask(b.loc[:,'dB'].lt(0),0).copy()
            timediff = np.array([])
            press = iter(range(len(b[b['dB']>0].index)-2))
            for j in press:
                timediff = np.append(timediff, b.loc[b[b['dB']>0].index[j+2],'Time'] - b.loc[b[b['dB']>0].index[j],'Time'])
                if (b.loc[b[b['dB']>0].index[j+2],'Time'] - b.loc[b[b['dB']>0].index[j],'Time'] <= 1) & (j < len((b[b['dB']>0].index))-5):
                    next(press)
                    next(press)
                    next(press)
            burstpress = np.append(burstpress, sum(timediff<=1))
            
    omittedburstpress = burstpress[omittedcues]/(15/60)
    rewardedburstpress = burstpress[rewardedcues]/(15/60)

    #maxspeed
    maxspeed = np.array([])
    for i in range(len(firstbar)):
        if barend[i] == 0:
            maxspeed = np.append(maxspeed, 0)
        else:
            maxspeed = np.append(maxspeed, a.loc[firstbar[i]:barend[i],'Speed'].max())

    omittedmaxspeed = maxspeed[omittedcues]
    rewardedmaxspeed = maxspeed[rewardedcues]
    
    

    omittedbar_exp = omittedbar_exp.append(pd.DataFrame(omittedbar).T)
    omittedbar_exp.index = animal_id

    rewardedbar_exp = rewardedbar_exp.append(pd.DataFrame(rewardedbar).T)
    rewardedbar_exp.index = animal_id
    
    
    omittedburstpress_exp = omittedburstpress_exp.append(pd.DataFrame(omittedburstpress).T)
    omittedburstpress_exp.index = animal_id
    
    rewardedburstpress_exp = rewardedburstpress_exp.append(pd.DataFrame(rewardedburstpress).T)
    rewardedburstpress_exp.index = animal_id

    
    omittedmaxspeed_exp = omittedmaxspeed_exp.append(pd.DataFrame(omittedmaxspeed).T)
    omittedmaxspeed_exp.index = animal_id
    
    rewardedmaxspeed_exp = rewardedmaxspeed_exp.append(pd.DataFrame(rewardedmaxspeed).T)
    rewardedmaxspeed_exp.index = animal_id
    

omittedbar_exp.columns += 1
rewardedbar_exp.columns += 1
omittedburstpress_exp.columns += 1
rewardedburstpress_exp.columns += 1
omittedmaxspeed_exp.columns += 1
rewardedmaxspeed_exp.columns += 1

with pd.ExcelWriter('expected_omission_by_cue.xlsx') as writer:
    omittedbar_exp.to_excel(writer, sheet_name = 'Bar Press Omitted')
    rewardedbar_exp.to_excel(writer, sheet_name = 'Bar Press Rewarded')
    omittedburstpress_exp.to_excel(writer, sheet_name = 'Burst Press Omitted')
    rewardedburstpress_exp.to_excel(writer, sheet_name = 'Burst Press Rewarded')
    omittedmaxspeed_exp.to_excel(writer, sheet_name = 'Max Speed Omitted')
    rewardedmaxspeed_exp.to_excel(writer, sheet_name = 'Max Speed Rewarded')

```
