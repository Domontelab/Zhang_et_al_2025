```python
import numpy as np
import scipy.io
import os
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
import seaborn as sb
import scipy.stats as stats
from sklearn.metrics import roc_auc_score as auROC
import sklearn.metrics as ms
```

```python
base_dir = 'C:/Users/NiE Microscope/Documents/MiniscopeV4/Minian/RawData/Satiation'
data_dir = os.path.abspath(os.path.join(base_dir, 'Spikes'))
session = ['Hungry','Satiated']

hungry_dir = os.path.abspath(os.path.join(data_dir, session[0]))
satiated_dir =  os.path.abspath(os.path.join(data_dir, session[1]))
crossreg_dir = os.path.abspath(os.path.join(base_dir, 'crossreg'))

os.listdir(hungry_dir)

FR = 20
CueN = 12
BinSize = 3
TimeRange = 10
Zsig = 1.96

FrameRange = FR*TimeRange
TotalFrame = 2*FrameRange
Bins = int(TotalFrame/BinSize)
```

# This section analyze all cells recorded in hungry or satiated session

```python
tracehungry = pd.DataFrame()


for n in range(len(os.listdir(hungry_dir))):
    f = pd.read_pickle(os.path.join(hungry_dir, os.listdir(hungry_dir)[n]))
    cueonset = np.array(np.where(np.diff(np.concatenate(([0],f.loc['cue']>0)))>0)[0])
    start = cueonset-FrameRange
    trace = f[:(len(f)-2)]
    tracenew = pd.DataFrame(0, index = trace.index, columns= np.arange(TotalFrame))
    for i in range(TotalFrame):
        tracenew.iloc[:,i] = trace.iloc[:,start+i].mean(axis=1)
    tracebin = tracenew.groupby(pd.cut(tracenew.columns,Bins),axis=1).mean()
    tracebin.columns = np.arange(0,Bins)
    tracezscore = tracebin.sub(tracebin.iloc[:,0:int((Bins/2))].mean(axis=1),axis = 0).divide(tracebin.iloc[:,0:int((Bins/2))].std(axis=1),axis=0)
    tracehungry = tracehungry.append(tracezscore)
    
tracesorted = tracehungry.sort_values(by=int((Bins/2)), axis =0)
ax = sb.heatmap(tracesorted, vmin = -4, vmax= 4, cmap = "bwr")
ax.set_xticks([0,int((Bins/2)),Bins])
ax.set_xticklabels([-TimeRange, 0, TimeRange])
print(str(len(tracehungry)) + ' neurons in Hungry')
```

# This section analyze the cells that can be matched across the two sessions

```python
tracehungry = pd.DataFrame()
animalid = pd.DataFrame()

for n in range(len(os.listdir(hungry_dir))):
    f = pd.read_pickle(os.path.join(hungry_dir, os.listdir(hungry_dir)[n]))
    animalid = pd.DataFrame(np.tile(n,len(f)-2), columns = ['animal'])
    cueonset = np.array(np.where(np.diff(np.concatenate(([0],f.loc['cue']>0)))>0)[0])
    start = cueonset-FrameRange
    trace = f[:(len(f)-2)] 
    tracenew = pd.DataFrame(0, index = trace.index, columns= np.arange(TotalFrame))
    for i in range(TotalFrame):
        tracenew.iloc[:,i] = trace.iloc[:,start+i].mean(axis=1)
    tracebin = tracenew.groupby(pd.cut(tracenew.columns,Bins),axis=1).mean()
    tracebin.columns = np.arange(0,Bins)
    tracezscore = tracebin.sub(tracebin.iloc[:,0:int((Bins/2))].mean(axis=1),axis = 0).divide(tracebin.iloc[:,0:int((Bins/2))].std(axis=1),axis=0)
    tracezscore.insert(0, 'animal', animalid.values)
    tracehungry = tracehungry.append(tracezscore)
```

```python
tracesatiated = pd.DataFrame()
animalid = pd.DataFrame()

for n in range(len(os.listdir(satiated_dir))):
    f = pd.read_pickle(os.path.join(satiated_dir, os.listdir(satiated_dir)[n]))
    animalid = pd.DataFrame(np.tile(n,len(f)-2), columns = ['animal'])
    cueonset = np.array(np.where(np.diff(np.concatenate(([0],f.loc['cue']>0)))>0)[0])
    start = cueonset-FrameRange
    trace = f[:(len(f)-2)] 
    tracenew = pd.DataFrame(0, index = trace.index, columns= np.arange(TotalFrame))
    for i in range(TotalFrame):
        tracenew.iloc[:,i] = trace.iloc[:,start+i].mean(axis=1)
    tracebin = tracenew.groupby(pd.cut(tracenew.columns,Bins),axis=1).mean()
    tracebin.columns = np.arange(0,Bins)
    tracezscore = tracebin.sub(tracebin.iloc[:,0:int((Bins/2))].mean(axis=1),axis = 0).divide(tracebin.iloc[:,0:int((Bins/2))].std(axis=1),axis=0)
    tracezscore.insert(0, 'animal', animalid.values)
    tracesatiated = tracesatiated.append(tracezscore)
```

```python
cross_hungry = pd.DataFrame()
cross_satiated = pd.DataFrame()

for i in range (len(os.listdir(crossreg_dir))):
    c = pd.read_csv(os.path.join(crossreg_dir, os.listdir(crossreg_dir)[i]), header = 1, index_col = 0, usecols = range(0,4), dtype = {'Hungry' : 'Int64', 'Satiated':'Int64'})
    c = c.dropna()
    chidx = np.array(c['Hungry'].values)
    csidx = np.array(c['Satiated'].values)

    hungry = tracehungry[tracehungry['animal']==i]
    hungryidx = np.array(hungry.index, dtype = int)

    satiated = tracesatiated[tracesatiated['animal']==i]
    satiatedidx = np.array(satiated.index, dtype = int)

    chidx = chidx[np.isin(chidx, hungryidx)]
    csidx = csidx[np.isin(csidx, satiatedidx)]

    c = c[np.isin(np.array(c['Hungry'].values, dtype =int), chidx)].reset_index(drop = True)
    c = c[np.isin(np.array(c['Satiated'].values, dtype =int), csidx)].reset_index(drop = True)

    hungry = hungry[np.isin(np.array(hungry.index, dtype=int), c['Hungry'].values)]
    satiated = satiated[np.isin(np.array(satiated.index, dtype=int), c['Satiated'].values)]

    hungry = hungry.reindex(index = np.array(c['Hungry'].values, dtype=str))
    satiated = satiated.reindex(index = np.array(c['Satiated'].values, dtype=str))

    hungry = hungry.reset_index()
    satiated = satiated.reset_index()

    cross_hungry = cross_hungry.append(hungry, ignore_index = True)
    cross_satiated = cross_satiated.append(satiated, ignore_index = True)

```

```python
cross_hungry_sorted = cross_hungry.sort_values(by=int((Bins/2)), axis =0)
ax = sb.heatmap(cross_hungry_sorted.iloc[:,2:], vmin = -4, vmax= 4, cmap = "bwr")
ax.set_xticks([0,int((Bins/2)),Bins])
ax.set_xticklabels([-TimeRange, 0, TimeRange])
print (str(len(cross_hungry)) +' neurons in Hungry ranked')
```

# auROC

```python
#before = cross_satiated.iloc[:,(int(Bins/2)-int(1*FR/BinSize)):(int((Bins/2)))].dropna(axis=0)
#after = cross_satiated.iloc[:,(int((Bins/2)+1)):(int(Bins/2)+int(1*FR/BinSize)+1)].dropna(axis=0)
before = cross_satiated.iloc[:,(int(Bins/2)+2-int(1*FR/BinSize)):(int((Bins/2)+2))].dropna(axis=0)
after = cross_satiated.iloc[:,(int((Bins/2)+2)):(int(Bins/2)+int(1*FR/BinSize)+2)].dropna(axis=0)

excitedmw = pd.DataFrame()
inhibitedmw = pd.DataFrame()
excitedindex = np.array([],dtype = int)
inhibitedindex = np.array([],dtype = int)

#stats.mannwhitneyu(after.iloc[2,:], before.iloc[2,:])
for n in range(len(cross_satiated.dropna(axis=0))):
    U, p = stats.mannwhitneyu(after.iloc[n,:], before.iloc[n,:])
    labels = np.concatenate((np.ones(after.iloc[n,:].shape), np.zeros(before.iloc[n,:].shape)))
    data = np.concatenate((after.iloc[n,:], before.iloc[n,:]))
    A = auROC(labels, data)
    if (A>0.5) & (p<0.01):
        excitedindex = np.append(excitedindex,n)
    elif (A<0.5) & (p<0.01):
        inhibitedindex = np.append(inhibitedindex,n)

excitedmw_satiated = cross_satiated.iloc[excitedindex,2::]
inhibitedmw_satiated = cross_satiated.iloc[inhibitedindex,2::]
excitedmw_satiated_inhungry = cross_hungry.iloc[excitedindex,2::]
inhibitedmw_satiated_inhungry = cross_hungry.iloc[inhibitedindex,2::]
```
