```python
from sklearn.preprocessing import MinMaxScaler

for animal_id in animal_id_list:
    # Ensure the DataFrame has enough columns
    if YDict[animal_id].shape[1] >= 8:
        # Split your data first
        X_train, X_test, y_train, y_test = train_test_split(
            XDict[animal_id], YDict[animal_id], test_size=0.2, random_state=2024
        )

        # Initialize scaler for the target variable
        scaler_train = MinMaxScaler()
        scaler_test = MinMaxScaler()

        # Select columns to scale in y (assuming these are continuous variables)
        columns_to_scale = y_train.columns[2:8]  # Adjust the columns as per your dataset specifics

        # Fit the scaler on the y_train and transform y_train and y_test
        for column in columns_to_scale:
            y_train[column] = scaler_train.fit_transform(y_train[[column]])  # Fit and transform on train
            y_test[column] = scaler_test.fit_transform(y_test[[column]])  # Transform test set with same scaler

        # Sorting by index (if necessary)
        X_train.sort_index(inplace=True)
        X_test.sort_index(inplace=True)
        y_train.sort_index(inplace=True)
        y_test.sort_index(inplace=True)

        # Update your dictionaries
        X_train_dict[animal_id] = X_train
        X_test_dict[animal_id] = X_test
        y_train_dict[animal_id] = y_train
        y_test_dict[animal_id] = y_test
```

```python
from sklearn.metrics import median_absolute_error


# Define decoding function with kNN decoder. For a simple demo, we will use the fixed number of neighbors 36.
def knn_decoding(embedding_train, 
                 embedding_test, 
                 label_train, 
                 label_test, 
                 n_neighbors=36):
    decoder = cebra.KNNDecoder(n_neighbors=n_neighbors, metric="cosine")

    decoder.fit(embedding_train, label_train)
    pred = decoder.predict(embedding_test)
    

    # Calculate z-scores of errors
    error_mean = np.mean(pred -label_test)
    error_std = np.std(pred - label_test)
    print(f"Range of maxmin {max(label_test)} : {min(label_test)}")
    
    mse = median_absolute_error(label_test, pred)
    normalized_errors = mse 
    




    # r2 of normalized errors
    r2_normalized = np.mean(normalized_errors)
        

    test_score = sklearn.metrics.r2_score(label_test, pred) 

    return test_score, r2_normalized, pred

def two_label_knn_decoding(embedding_train, 
                                embedding_test, 
                                label_train,
                                label_test, 
                                n_neighbors=36):
    approach_decoder = cebra.KNNDecoder(n_neighbors=n_neighbors, metric="cosine")
    lever_decoder = cebra.KNNDecoder(n_neighbors=n_neighbors, metric="cosine")

    approach_decoder.fit(embedding_train, label_train[:,0])
    lever_decoder.fit(embedding_train, label_train[:,1])

    approach_pred = approach_decoder.predict(embedding_test)
    lever_pred = lever_decoder.predict(embedding_test)

    prediction = np.stack([approach_pred, lever_pred],axis = 1)

    test_score = sklearn.metrics.r2_score(label_test[:,:2], prediction)
    
    error_mean_approach = np.mean(prediction[:,0] - label_test[:, 0])
    error_std_approach = np.std(prediction[:,0] - label_test[:, 0])
    error_mean_lever = np.mean(prediction[:,1] - label_test[:, 1])
    error_std_lever = np.std(prediction[:,1] - label_test[:, 1])

    mse_approach = median_absolute_error(label_test[:, 0], prediction[:,0])
    normalized_errors_approach = mse_approach
    
    mse_lever = median_absolute_error(label_test[:, 1], prediction[:,1])
    normalized_errors_lever = mse_lever 
    
    approach_test_score = sklearn.metrics.r2_score(label_test[:, 0], prediction[:,0])
    
    # r2 of normalized errors
    r2_normalized_approach = np.mean(normalized_errors_approach)
    r2_normalized_lever = np.mean(normalized_errors_lever)
    r2_normalized = np.mean([r2_normalized_approach, r2_normalized_lever])

    
    return test_score, r2_normalized, prediction, approach_test_score
```

```python
# Initialize decoding dictionaries
freeze_hidden_decode_dict = {}
freeze_hidden_shuffled_decode_dict = {}
approach_lever_decode_dict = {}
approach_lever_shuffled_decode_dict = {}
tone_decode_dict = {}
tone_shuffled_decode_dict = {}
speed_decode_dict = {}
speed_shuffled_decode_dict = {}


for animal_id in animal_id_list:
    # Freezing, In Hidden Area
    cebra_freeze_hidden_train = y_train_dict[animal_id].iloc[:, [3, 5]].to_numpy()
    cebra_freeze_hidden_test = y_test_dict[animal_id].iloc[:, [3, 5]].to_numpy()

    cebra_freeze_hidden_embedding_train = cebra_behavior_model_comb1_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_freeze_hidden_embedding_test = cebra_behavior_model_comb1_dict[animal_id].transform(X_test_dict[animal_id])
    cebra_freeze_hidden_shuffled_embedding_train = cebra_behavior_model_comb1_shuffled_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_freeze_hidden_shuffled_embedding_test = cebra_behavior_model_comb1_shuffled_dict[animal_id].transform(X_test_dict[animal_id])

    freeze_hidden_decode_dict[animal_id] = two_label_knn_decoding(cebra_freeze_hidden_embedding_train, 
                                                              cebra_freeze_hidden_embedding_test, 
                                                              cebra_freeze_hidden_train, 
                                                              cebra_freeze_hidden_test)

    freeze_hidden_shuffled_decode_dict[animal_id] = two_label_knn_decoding(cebra_freeze_hidden_shuffled_embedding_train,
                                                                       cebra_freeze_hidden_shuffled_embedding_test, 
                                                                       cebra_freeze_hidden_train, 
                                                                       cebra_freeze_hidden_test)

    
    # Approach, Lever
    cebra_approach_lever_train = y_train_dict[animal_id].iloc[:, [4, 6]].to_numpy()
    cebra_approach_lever_test = y_test_dict[animal_id].iloc[:, [4, 6]].to_numpy()

    cebra_approach_lever_embedding_train = cebra_behavior_model_comb2_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_approach_lever_embedding_test = cebra_behavior_model_comb2_dict[animal_id].transform(X_test_dict[animal_id])
    cebra_approach_lever_shuffled_embedding_train = cebra_behavior_model_comb2_shuffled_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_approach_lever_shuffled_embedding_test = cebra_behavior_model_comb2_shuffled_dict[animal_id].transform(X_test_dict[animal_id])

    approach_lever_decode_dict[animal_id] = two_label_knn_decoding(cebra_approach_lever_embedding_train, 
                                                              cebra_approach_lever_embedding_test, 
                                                              cebra_approach_lever_train, 
                                                              cebra_approach_lever_test)

    approach_lever_shuffled_decode_dict[animal_id] = two_label_knn_decoding(cebra_approach_lever_shuffled_embedding_train,
                                                                       cebra_approach_lever_shuffled_embedding_test, 
                                                                       cebra_approach_lever_train, 
                                                                       cebra_approach_lever_test)

    # Tone Active
    print(animal_id)
    cebra_tone_label_train = y_train_dict[animal_id].iloc[:, [8]].to_numpy()
    cebra_tone_label_test = y_test_dict[animal_id].iloc[:, [8]].to_numpy()

    cebra_tone_embedding_train = cebra_behavior_model_comb3_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_tone_embedding_test = cebra_behavior_model_comb3_dict[animal_id].transform(X_test_dict[animal_id])
    cebra_tone_shuffled_embedding_train = cebra_behavior_model_comb3_shuffled_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_tone_shuffled_embedding_test = cebra_behavior_model_comb3_shuffled_dict[animal_id].transform(X_test_dict[animal_id])

    tone_decode_dict[animal_id] = knn_decoding(cebra_tone_embedding_train, 
                                      cebra_tone_embedding_test, 
                                      cebra_tone_label_train, 
                                      cebra_tone_label_test)

    tone_shuffled_decode_dict[animal_id] = knn_decoding(cebra_tone_shuffled_embedding_train, 
                                      cebra_tone_shuffled_embedding_test, 
                                      cebra_tone_label_train, 
                                      cebra_tone_label_test)


    # Speed
    cebra_speed_label_train = y_train_dict[animal_id].iloc[:, [2]].to_numpy()
    cebra_speed_label_test = y_test_dict[animal_id].iloc[:, [2]].to_numpy()

    cebra_speed_embedding_train = cebra_behavior_model_comb4_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_speed_embedding_test = cebra_behavior_model_comb4_dict[animal_id].transform(X_test_dict[animal_id])
    cebra_speed_shuffled_embedding_train = cebra_behavior_model_comb4_shuffled_dict[animal_id].transform(X_train_dict[animal_id])
    cebra_speed_shuffled_embedding_test = cebra_behavior_model_comb4_shuffled_dict[animal_id].transform(X_test_dict[animal_id])

    speed_decode_dict[animal_id] = knn_decoding(cebra_speed_embedding_train, 
                                      cebra_speed_embedding_test, 
                                      cebra_speed_label_train, 
                                      cebra_speed_label_test)

    speed_shuffled_decode_dict[animal_id] = knn_decoding(cebra_speed_shuffled_embedding_train, 
                                      cebra_speed_shuffled_embedding_test, 
                                      cebra_speed_label_train, 
                                      cebra_speed_label_test)



```

```python
def get_predictions_and_errors_one_dim(embedding_train, embedding_test, labels_train, labels_test, n_neighbors=36, metric="cosine"):
    """Perform KNN decoding and calculate prediction errors for one-dimensional labels."""
    decoder = cebra.KNNDecoder(n_neighbors=n_neighbors, metric=metric)
    decoder.fit(embedding_train, labels_train)
    predictions = decoder.predict(embedding_test)
    errors = np.abs(predictions - labels_test)
    return errors

def get_predictions_and_errors_two_dim(embedding_train, embedding_test, labels_train, labels_test, n_neighbors=36):
    """Perform KNN decoding for two-dimensional labels and calculate average prediction errors."""
    approach_decoder = cebra.KNNDecoder(n_neighbors=n_neighbors, metric="cosine")
    lever_decoder = cebra.KNNDecoder(n_neighbors=n_neighbors, metric="cosine")

    approach_decoder.fit(embedding_train, labels_train[:, 0])
    lever_decoder.fit(embedding_train, labels_train[:, 1])

    approach_pred = approach_decoder.predict(embedding_test)
    lever_pred = lever_decoder.predict(embedding_test)

    prediction = np.stack([approach_pred, lever_pred], axis=1)
    errors = np.mean(np.abs(prediction - labels_test), axis=1)
    
    return errors

def calculate_ecdf(data):
    """Calculate the ECDF for a given 1D array."""
    sorted_data = np.sort(data.flatten())
    ecdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    return sorted_data, ecdf

def calculate_average_sem(ecdf_dict):
    """Calculate average and SEM for ECDFs stored in a dictionary."""
    all_sorted_errors = []
    all_ecdfs = []

    for animal_id in ecdf_dict:
        sorted_errors, ecdf = ecdf_dict[animal_id]
        all_sorted_errors.append(sorted_errors)
        all_ecdfs.append(ecdf)

    # Find the longest sorted_errors array for consistent x-axis
    max_len = max(len(errors) for errors in all_sorted_errors)

    # Extend all arrays to the max length by padding with the last value
    extended_errors = np.array([np.pad(errors, (0, max_len - len(errors)), 'edge') for errors in all_sorted_errors])
    extended_ecdfs = np.array([np.pad(ecdf, (0, max_len - len(ecdf)), 'edge') for ecdf in all_ecdfs])

    average_ecdf = np.mean(extended_ecdfs, axis=0)
    sem_ecdf = sem(extended_errors, axis=0)

    return extended_errors[0], average_ecdf, sem_ecdf

# Initialize storage for ECDFs
ecdf_speed_dict = {}
ecdf_speed_shuffled_dict = {}
ecdf_approach_lever_dict = {}
ecdf_approach_lever_shuffled_dict = {}
ecdf_freezing_hidden_dict = {}
ecdf_freezing_hidden_shuffled_dict = {}
ecdf_tone_dict = {}
ecdf_tone_shuffled_dict = {}

for animal_id in animal_id_list:
    # Speed original embedding
    speed_errors = get_predictions_and_errors_one_dim(
        cebra_behavior_model_comb4_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb4_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [2]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [2]].to_numpy()
    )
    ecdf_speed_dict[animal_id] = calculate_ecdf(speed_errors)

    # Speed shuffled embedding
    speed_shuffled_errors = get_predictions_and_errors_one_dim(
        cebra_behavior_model_comb4_shuffled_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb4_shuffled_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [2]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [2]].to_numpy()
    )
    ecdf_speed_shuffled_dict[animal_id] = calculate_ecdf(speed_shuffled_errors)
    
    # Approach Lever original embedding
    approach_lever_errors = get_predictions_and_errors_two_dim(
        cebra_behavior_model_comb2_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb2_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [4, 6]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [4, 6]].to_numpy()
    )
    ecdf_approach_lever_dict[animal_id] = calculate_ecdf(approach_lever_errors)

    # Approach Lever shuffled embedding
    approach_lever_shuffled_errors = get_predictions_and_errors_two_dim(
        cebra_behavior_model_comb2_shuffled_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb2_shuffled_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [4, 6]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [4, 6]].to_numpy()
    )
    ecdf_approach_lever_shuffled_dict[animal_id] = calculate_ecdf(approach_lever_shuffled_errors)
    
    # Freezing Hidden original embedding
    freezing_hidden_errors = get_predictions_and_errors_two_dim(
        cebra_behavior_model_comb1_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb1_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [3, 5]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [3, 5]].to_numpy()
    )
    ecdf_freezing_hidden_dict[animal_id] = calculate_ecdf(freezing_hidden_errors)

    # Freezing Hidden shuffled embedding
    freezing_hidden_shuffled_errors = get_predictions_and_errors_two_dim(
        cebra_behavior_model_comb1_shuffled_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb1_shuffled_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [3, 5]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [3, 5]].to_numpy()
    )
    ecdf_freezing_hidden_shuffled_dict[animal_id] = calculate_ecdf(freezing_hidden_shuffled_errors)

    # Tone original embedding
    tone_errors = get_predictions_and_errors_one_dim(
        cebra_behavior_model_comb3_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb3_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [8]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [8]].to_numpy()
    )
    ecdf_tone_dict[animal_id] = calculate_ecdf(tone_errors)

    # Tone shuffled embedding
    tone_shuffled_errors = get_predictions_and_errors_one_dim(
        cebra_behavior_model_comb3_shuffled_dict[animal_id].transform(X_train_dict[animal_id]),
        cebra_behavior_model_comb3_shuffled_dict[animal_id].transform(X_test_dict[animal_id]),
        y_train_dict[animal_id].iloc[:, [8]].to_numpy(),
        y_test_dict[animal_id].iloc[:, [8]].to_numpy()
    )
    ecdf_tone_shuffled_dict[animal_id] = calculate_ecdf(tone_shuffled_errors)

# Calculate average and SEM ECDF for each outcome
sorted_errors_speed_avg, ecdf_speed_avg, ecdf_speed_sem = calculate_average_sem(ecdf_speed_dict)
sorted_errors_speed_shuffled_avg, ecdf_speed_shuffled_avg, ecdf_speed_shuffled_sem = calculate_average_sem(ecdf_speed_shuffled_dict)
sorted_errors_approach_lever_avg, ecdf_approach_lever_avg, ecdf_approach_lever_sem = calculate_average_sem(ecdf_approach_lever_dict)
sorted_errors_approach_lever_shuffled_avg, ecdf_approach_lever_shuffled_avg, ecdf_approach_lever_shuffled_sem = calculate_average_sem(ecdf_approach_lever_shuffled_dict)
sorted_errors_freezing_hidden_avg, ecdf_freezing_hidden_avg, ecdf_freezing_hidden_sem = calculate_average_sem(ecdf_freezing_hidden_dict)
sorted_errors_freezing_hidden_shuffled_avg, ecdf_freezing_hidden_shuffled_avg, ecdf_freezing_hidden_shuffled_sem = calculate_average_sem(ecdf_freezing_hidden_shuffled_dict)
sorted_errors_tone_avg, ecdf_tone_avg, ecdf_tone_sem = calculate_average_sem(ecdf_tone_dict)
sorted_errors_tone_shuffled_avg, ecdf_tone_shuffled_avg, ecdf_tone_shuffled_sem = calculate_average_sem(ecdf_tone_shuffled_dict)

```

```python
def plot_binary_spans(ax, timestamps, binary_values, color, ylabel, title = None, bottomspine = None, xlabel = None, xtick=None, alpha=0.8):
    """
    Plot horizontal spans for periods where binary_values are 1, without assuming timestamp format.

    Parameters:
    - timestamps: Sequence of any sequential identifiers.
    - binary_values: List, array, or pandas Series of binary values (0s and 1s).
    - color: Color of the spans. Default is 'blue'.
    - alpha: Transparency of the spans. Default is 0.5.
    """

    # Convert binary_values to a list if it's a pandas Series
    if isinstance(binary_values, pd.Series):
        binary_values_list = binary_values.tolist()
    else:
        binary_values_list = binary_values

    start_colored_span = None
    for i, value in enumerate(binary_values_list):
        if value > 0.2 and start_colored_span is None:
            # Mark the start of a span
            start_colored_span = i
        elif value <= 0.2 and start_colored_span is not None:
            # End of a span, add it to the plot
            ax.axvspan(timestamps[start_colored_span], timestamps[i], facecolor=color, alpha=alpha, edgecolor = 'none')
            start_colored_span = None
    
    # Check if the last value in the series is '1' to close the last span
    if binary_values_list[-1] > 0.2 and start_colored_span is not None :
        ax.axvspan(timestamps[start_colored_span], timestamps[-1], facecolor=color, alpha=alpha, edgecolor = 'none')

    ax.set_xlim(0, max(timestamps))
    ax.set_ylabel(ylabel, rotation = 0, verticalalignment = 'center', fontsize = 18, horizontalalignment = 'right')
    if title is not None:
        plt.title(title)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    if bottomspine is None:
        ax.spines['bottom'].set_visible(False)
        
    if xlabel is not None:
        ax.set_xlabel(xlabel, fontsize = 18)
        ax.tick_params(labelsize=16)
        
    if xtick is None:
        ax.set_xticks([])
    
    ax.set_yticks([])
    



fig, axs = plt.subplots(3, 1, figsize=(10, 3), dpi = 300)

# In Food Area, Bar Press Active
labels = Y.iloc[:, [3, 5]].to_numpy()
test_labels = y_test.iloc[:, [3, 5]].to_numpy()
pred = freeze_hidden_decode_dict[animal_id][2]
pred_shuffle = freeze_hidden_shuffled_decode_dict[animal_id][2]


colors = sns.color_palette('Paired')

fig.suptitle('Freezing', fontsize=16)



plot_binary_spans(axs[0], y_test.Time.values, y_test['Freezing'].values, color = colors[0], ylabel = 'Ground Truth')
plot_binary_spans(axs[1],y_test.Time.values, pred[:, 0], color = colors[0], ylabel = 'Predicted')
plot_binary_spans(axs[2],y_test.Time.values, pred_shuffle[:,0], color = 'gray', ylabel = 'Shuffled', title = None, bottomspine = True, xlabel = 'Time(s)', xtick=True)


plt.tight_layout()
plt.savefig("/scratch/09117/xz6783/Xu/Plots/Fig2/fig2_catodor_pred_freezing.svg", transparent=True)
plt.show()
```

```python
def plot_binary_spans(ax, timestamps, binary_values, color, ylabel, title = None, bottomspine = None, xlabel = None, xtick=None, alpha=0.8):
    """
    Plot horizontal spans for periods where binary_values are 1, without assuming timestamp format.

    Parameters:
    - timestamps: Sequence of any sequential identifiers.
    - binary_values: List, array, or pandas Series of binary values (0s and 1s).
    - color: Color of the spans. Default is 'blue'.
    - alpha: Transparency of the spans. Default is 0.5.
    """

    # Convert binary_values to a list if it's a pandas Series
    if isinstance(binary_values, pd.Series):
        binary_values_list = binary_values.tolist()
    else:
        binary_values_list = binary_values

    start_colored_span = None
    for i, value in enumerate(binary_values_list):
        if value > 0.2 and start_colored_span is None:
            # Mark the start of a span
            start_colored_span = i
        elif value <= 0.2 and start_colored_span is not None:
            # End of a span, add it to the plot
            ax.axvspan(timestamps[start_colored_span], timestamps[i], facecolor=color, alpha=alpha, edgecolor = 'none')
            start_colored_span = None
    
    # Check if the last value in the series is '1' to close the last span
    if binary_values_list[-1] > 0.2 and start_colored_span is not None :
        ax.axvspan(timestamps[start_colored_span], timestamps[-1], facecolor=color, alpha=alpha, edgecolor = 'none')

    ax.set_xlim(0, max(timestamps))
    ax.set_ylabel(ylabel, rotation = 0, verticalalignment = 'center', fontsize = 18, horizontalalignment = 'right')
    if title is not None:
        plt.title(title)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    if bottomspine is None:
        ax.spines['bottom'].set_visible(False)
        
    if xlabel is not None:
        ax.set_xlabel(xlabel, fontsize = 18)
        ax.tick_params(labelsize=16)
        
    if xtick is None:
        ax.set_xticks([])
    
    ax.set_yticks([])
    



fig, axs = plt.subplots(3, 1, figsize=(10, 3), dpi = 300)

# In Food Area, Bar Press Active
labels = Y.iloc[:, [8]].to_numpy()
test_labels = y_test.iloc[:, [8]].to_numpy()
pred = tone_decode_dict[animal_id][2]
pred_shuffle = tone_shuffled_decode_dict[animal_id][2]



colors = sns.color_palette('Paired')

fig.suptitle('Tone', fontsize=16)



plot_binary_spans(axs[0], y_test.Time.values, y_test['Tone active'].values, color = colors[1], ylabel = 'Ground Truth')
plot_binary_spans(axs[1],y_test.Time.values, pred, color = colors[1], ylabel = 'Predicted')
plot_binary_spans(axs[2],y_test.Time.values, pred_shuffle, color = 'gray', ylabel = 'Shuffled', title = None, bottomspine = True, xlabel = 'Time(s)', xtick=True)


plt.tight_layout()
plt.savefig("/scratch/09117/xz6783/Xu/Plots/Fig2/fig2_catodor_pred_tone.svg", transparent=True)
plt.show()
```
